````python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
````

### customize a dataset

````python
data = {
    'ID':[1,2,3,4,5,6],
    'Onion':[1,0,0,1,1,1],
    'Potato':[1,1,0,1,1,1],
    'Burger':[1,1,0,0,1,1],
    'Milk':[0,1,1,1,0,1],
    'Beer':[0,0,1,0,1,0]
}
````

````python
df = pd.DataFrame(data)
````

````python
df = df[['ID', 'Onion', 'Potato', 'Burger', 'Milk', 'Beer' ]]
````

````python
df
````

### Filter with support to select frequent itemsets
- min support = 50%
- `apriori(df, min_support=0.5, use_colnames=True)`

````python
# http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/
frequent_itemsets = apriori(df[['Onion', 'Potato','Burger', 'Milk', 'Beer']], min_support=0.5, use_colnames=True)
````

````python
frequent_itemsets
````

### Association Rules
- ```association_rules(df, metric='lift', min_threshold=1)```

````python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
````

````python
rules
````

````python
rules [ (rules['lift'] >1.125)  & (rules['confidence']> 0.8)  ]
````

### Playing with MovieLens

````python
movies = pd.read_csv('ml-latest-small/movies.csv')
````

````python
movies.head(10)
````

````python
movies_ohe = movies.drop('genres', 1).join(movies.genres.str.get_dummies())
````

````python
pd.options.display.max_columns = 100
````

````python
movies_ohe.head()
````

````python
# 9125 pieces of data, 22-2=20 classes
movies_ohe.shape
````

````python
# set movieId and title column as index
movies_ohe.set_index(['movieId', 'title'], inplace=True)
````

````python
movies_ohe.head()
````

````python
frequent_itemsets_movies = apriori(movies_ohe, use_colnames=True, min_support=0.025)
````

````python
frequent_itemsets_movies
````

````python
rule_movies = association_rules(frequent_itemsets_movies, metric='lift', min_threshold=1.25)
````

````python
rule_movies
````

````python
rule_movies[(rule_movies.lift>4)].sort_values(by=['lift'], ascending=False)
````
